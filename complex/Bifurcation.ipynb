```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r): # Logistic map function
    return r * x * (1 - x)

r_values = [2.8, 3.2, 3.5]   # r values
colors = ['b', 'g', 'r']   # colors for each r value
labels = [r'$r = 2.8$ (Period 1)', r'$r = 3.2$ (Period 2)', r'$r = 3.5$ (Period 4)']
initial_conditions = np.linspace(0.1, 0.9, 10)      # initial conditions

plt.figure(figsize=(4, 4) , dpi=150) # Create a new figure I chose dpi=150. can change if needed

for r, color, label in zip(r_values, colors, labels): # Loop over r values
    for x0 in initial_conditions:
        x_values = [x0]
        for whatever in range(100):     # Iterate the logistic map function 100 times
            x_values.append(logistic_map(x_values[-1], r)) # Append the new value to the list
        
        plt.plot(x_values[:-1], x_values[1:], 'o', markersize=2, color=color, alpha=0.5, label=label if x0 == initial_conditions[0] else "")

plt.xlabel("$x_n$")
plt.ylabel("$x_{n+1}$")
plt.title("Phase Space Plot for various $r$ values")
plt.legend()
plt.grid(True)

plt.show()
```



### This is the Bifurcation diagaram, which is a plot of the stable values of $x_n$ as a function of r. 
I will plot the last 200 iterations for each r in order to get a sense of the stable values of $x_n$ for each r.



```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r): # Logistic map function
    return r * x * (1 - x) 


r_values = np.linspace(0, 4, 2000) # r values to iterate over
num_iterations = 1000  # Number of iterations for each r value
last_iterations = 200  # Number of iterations to keep
x0 = 0.5  # Initial condition


plt.figure(figsize=(10, 6)) # Create a new figure

for r in r_values:
    x = x0
    for whatever in range(num_iterations - last_iterations):   # Iterate to discard transient
        x = logistic_map(x, r) # Iterate the logistic map
    x_values = [] # List to store last iterations
    for whatever in range(last_iterations):  # Iterate to store last iterations
        x = logistic_map(x, r) # Iterate the logistic map
        x_values.append(x) # Append the current x value to the list
    plt.plot([r] * last_iterations, x_values, ',b', alpha=0.5)  # Plot the last iterations

plt.xlabel("r")
plt.ylabel("x")
plt.title("Bifurcation of the Logistic Map")
plt.xlim(0, 4)
plt.ylim(0, 1)
plt.grid(True)
plt.savefig("bifurcation_diagram.png")

plt.show()
```
