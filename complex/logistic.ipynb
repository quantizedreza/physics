```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the logistic growth equation
def logistic_growth(x, t, r, k):
    return r * x * (1 - x / k)

# Parameters
r = 0.3  # Growth rate
k = 0.8  # Carrying capacity

# Time points
t = np.linspace(0, 20, 1000)  # From 0 to 20 time units with 1000 points

# Initial condition
x0 = 0.1  # Start with small population

# Solve the ODE
x = odeint(logistic_growth, x0, t, args=(r, k))

# Plot the result
plt.figure(figsize=(10, 6))
plt.plot(t, x, label=f'Logistic Growth (r={r}, k={k})')
plt.xlabel('Time')
plt.ylabel('Population (x)')
plt.title('Logistic Growth Model')
plt.legend()
plt.grid(True)
plt.savefig('logistic_growth_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```
